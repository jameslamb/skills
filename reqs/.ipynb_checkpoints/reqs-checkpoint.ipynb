# Parsing Job Descriptions

## Approach 1: Web Scraping Job Descriptions

As of late November, I've got all of the posted jobs for Uber, Square, and Affirm as text files through a scraping process that's contained in the files in this directory.

The goal is to parse out all of the skills associated with each role.

Immediate use case: figure out which skills to ask candidates about. Also aggregate the data to show what the highest-demand skills/traits are, by frequency.

Longer-term use cases:
- If we can reliably scrape, then we can turn this into an up-to-the-day-accurate D3 viz on our site for a variety of companies. "What skills are in demand?"
- Database of jobs, current and previous, at... all of the companies out here, and the skills associated with them. 

### Challenges
- Not all JDs have an "about Uber" section, so it's hard to know what to cut. So perhaps just specify the lines that I _don't_ want included, because I know they're fluff.

### Notes
- It may be worth using a tool for deduplication and entity resolution to process the words (OpenRefine? http://openrefine.org/)

```python
# Start out with one arbitrary job; 
# when this code is working, loop through all of them

throwaway_lines = ['we’re chan', 'about uber', 'for the pe', 'and that’s', 'about the ',
                  'what you’l', 'bonus poin', 'see our ca', 'at uber we']

digits = [item+'\n' for item in string.digits]

list_items = []
titles = []

with open('uber/concat_uber_jobs.txt', 'r') as infile:
    for index, line in enumerate(infile):
        if index == 0:
            title = line
        elif line.strip() == '############ LIST_CONTENTS ############':
            continue
        elif line[:10].lower() in throwaway_lines:
            continue
        elif line.strip() in ['None', 'None ', 'None \n', 'None\n', '\n',' \n' '\n ', '  ', '',
                      '############ START_LIST_CONTENTS ############']:
            continue
        elif line in digits:
            continue
        else:
            list_items.append(line.strip())
        titles.append(title)

# for item in list_items:
#     print(item)
# pprint(list_items)
```

```python
sw = get_stop_words('en')

word_bank = []

for item in list_items:
    words = re.findall(r'\w+', item)
    # Remove stopwords, remove single letters, remove single numbers
    quality_words = [i for i in words if i not in sw and i not in string.ascii_letters and i not in string.digits]
    # Lemmatize the words to standardize them
    for word in quality_words:
        word_bank.append(word.lower())
        
print("Sample Word Bank:\n\n")
print(word_bank[:20])
```

```python
print("Current stop words:\n\n", sw)
```

```python
# There appear to be errors with lemmatizing; dropped that version

clean_word_bank = []

for w in word_bank:
    word = Word(w)
    clean_word_bank.append(word.lemmatize())
    
differences = [word for word in clean_word_bank if word not in word_bank]

print("Sample of differences between lemmatized and non-lemmatized output:\n\n")
print(differences[:20])
```

```python
data = word_bank

word_dict = {}

for word in data:
    if not word in word_dict:
        word_dict[word] = 1
    else:
        word_dict[word] += 1
        
# print(word_dict)

keys = word_dict.keys()
vals = word_dict.values()

zipped = sorted(zip(vals,keys), reverse=True)

keys_sorted = [keys for vals, keys in zipped]
vals_sorted = [vals for vals, keys in zipped]

print("Sample Zipped Sorted Values:\n\n")
print(zipped[:20])

print("\n\nSample Sorted Keys:\n\n")
print(keys_sorted[:20])

print("\n\nSample Sorted Values:\n\n")
print(vals_sorted[:20])
```

```python
def make_list(infile):
    mylist = []
    with open(infile, 'r') as f:
        for line in f:
            line = line.strip()
            word, count = line.split(',')
            mylist.append(word)
    return(mylist)
            
square = make_list('square_words.csv')
affirm = make_list('affirm_words.csv')
uber = make_list('uber_words.csv')

all_lists = square + affirm + uber

combined_scraped = list(set(square + affirm + uber))
# combined_scraped = list(set(combined_scraped))

print("Non-unique: ", len(all_lists))
print("Unique words: ", len(combined_scraped))
```

## Find n-grams from combined documents
_Go through each corpus and remove the "fluff" lines ("Here at Uber, we believe that..."). We're only interested in the lines that describe jobs. The process is slightly different for each company -- which is why this section is verbose._

```python
# Uber

uber_throwaway_lines = ['we’re chan', 'about uber', 'for the pe', 'and that’s', 'about the ',
                  'what you’l', 'bonus poin', 'see our ca', 'at uber we']

digits = [item+'\n' for item in string.digits]

uber_meat = []

with open('uber/concat_uber_jobs.txt', 'r') as infile:
    for index, line in enumerate(infile):
        if index == 0:
            title = line
        elif line.strip() == '############ LIST_CONTENTS ############':
            continue
        elif line[:10].lower() in uber_throwaway_lines:
            continue
        elif line.strip() in ['None', 'None ', 'None \n', 'None\n', '\n',' \n' '\n ', '  ', '',
                      '############ START_LIST_CONTENTS ############']:
            continue
        elif line in digits:
            continue
        else:
            uber_meat.append(line.strip())
        
# Affirm

affirm_throwaway_lines = ['affirm hom', 'jobs power']

affirm_meat = []

with open('affirm/concat_affirm_jobs.txt', 'r') as infile:
    for index, line in enumerate(infile):
        if index == 0:
            title = line
        elif line.strip() == '############ LIST_CONTENTS ############':
            list_meat is True
        elif line[:10].lower() in affirm_throwaway_lines:
            continue
        elif line.strip() in ['None', 'None ', 'None \n', 'None\n', '\n',' \n' '\n ', '  ', '',
                      '############ START_LIST_CONTENTS ############']:
            continue
        elif line in digits:
            continue
        else:
            # Specific to Affirm; 'Jobs powered by' gets dropped into JDs
            line = line.replace('Jobs powered by', '')
            affirm_meat.append(line.strip())
        
# Square
        
square_throwaway_lines = ['we believe', 'at square,']

square_meat = []

with open('square/concat_square_jobs.txt', 'r') as infile:
    for index, line in enumerate(infile):
        if index == 0:
            title = line
        elif line.strip() == '############ LIST_CONTENTS ############':
            continue
        elif line.strip() == '############ SPAN ############':
            continue
        elif line[:10].lower() in square_throwaway_lines:
            continue
        elif line.strip() in ['None', 'None ', 'None \n', 'None\n', '\n',' \n' '\n ', '  ', '',
                      '############ START_LIST_CONTENTS ############', 'Full-time', 'You have:']:
            continue
        elif line in digits:
            continue
        else:
            if line.strip() not in square_meat:
                square_meat.append(line.strip())
                
print("Output\n")
print("Square:\n", square_meat[:3])
print("Affirm:\n", affirm_meat[:3])
print("Uber:\n", uber_meat[:3])
```

```python
all_job_text = affirm_meat + square_meat + uber_meat

with open('combined_filtered_jd_lines.txt', 'w+') as outfile:
    for line in all_job_text:
        outfile.write(line)
        outfile.write("\n")
```

```python
WORD_RE = re.compile(r"[\w']+", re.UNICODE)

DEFAULT_STOP_WORDS = [
    'a', 'about', 'also', 'am', 'an', 'and', 'any', 'are', 'as', 'at', 'be',
    'but', 'by', 'can', 'com', 'did', 'do', 'does', 'for', 'from', 'had',
    'has', 'have', 'he', "he'd", "he'll", "he's", 'her', 'here', 'hers',
    'him', 'his', 'i', "i'd", "i'll", "i'm", "i've", 'if', 'in', 'into', 'is',
    'it', "it's", 'its', 'just', 'me', 'mine', 'my', 'of', 'on', 'or', 'org',
    'our', 'ours', 'she', "she'd", "she'll", "she's", 'some', 'than', 'that',
    'the', 'their', 'them', 'then', 'there', 'these', 'they', "they'd",
    "they'll", "they're", 'this', 'those', 'to', 'us', 'was', 'we', "we'd",
    "we'll", "we're", 'were', 'what', 'where', 'which', 'who', 'will', 'with',
    'would', 'you', 'your', 'yours']

nonsense_words = ['re', 'as, ''ll', 'na', 'da', 'etc', 'que', 'things', 'ou', 'co', 'para',
                 'parceiros', 'com', 've', 'ou', 'nossos', 'experiência', 'um', 'ser', 'l0',
                 'se', 'são', 'ajudar','você', 'ter', 'uma', 'don','por', 'não', 'an', 'emea',
                 'dos', 'sua', 'pela', 'eles','apac', 'trabalhar', 'sabe', 's', 'e', 'g']

all_stop_words = DEFAULT_STOP_WORDS + nonsense_words
```

### Two-grams

```python
from nltk import ngrams

twogram_dict = {}

words = [word.lower() for word in WORD_RE.findall(all_text_no_apostrophes)
         if word.lower() not in all_stop_words]

all_words = ' '.join(words)

print(all_words[:2000],"\n\n")


n = 2
twograms = ngrams(all_words.split(), n)

for gram in twograms:
    if gram in twogram_dict:
        twogram_dict[gram] += 1
    else:
        twogram_dict[gram] = 1
```

### Three-grams

```python

def sorted_ngram_dict(n, in_string):
    WORD_RE = re.compile(r"[\w']+", re.UNICODE)
    mydict = {}
    my_grams = ngrams(in_string.split(), n)
    for gram in my_grams:
        if gram in mydict:
            mydict[gram] += 1
        else:
            mydict[gram] = 1
    # Sort the dictionary and turn each key into one string without quotes
    s = [(' '.join(WORD_RE.findall(str(k).replace("'",''))), 
          mydict[k]) for k in sorted(mydict, key=mydict.get, reverse=True)]
    return s
        
threegram_dict = sorted_ngram_dict(3, all_words)
    
counter = 0

for key, value in threegram_dict:
    if counter < 30:
        counter += 1
        print(key, "\t", value)
    
# Write threegrams to file
with open('threegrams.txt', 'w+') as outfile:
    for key, value in threegram_dict:
        outfile.write(key)
        outfile.write("\t")
        outfile.write(str(value))
        outfile.write("\n")
```
